```python
# Import Libraries
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

```python
# Import US Data
us_videos = pd.read_csv('data/USvideos.csv')
us_videos_categories = pd.read_json('data/US_category_id.json')
# Map Category IDs using the supporting file: US_category_id.json
categories = [{"id":int(category['id']), "category":category['snippet']['title']} for category in us_videos_categories['items']]
categories
```

```python
us_videos.head()
```

```python
# First Row
# us_videos = us_videos.drop(columns = ['video_id','publishedAt','tags','thumbnail_link','comments_disabled','ratings_disabled','description'])

us_videos.index.name = 'id'
us_videos.head()
```

```python
# Create dataframe for category ID
cat_df = pd.DataFrame(categories)
cat_df.set_index("id",inplace=True)
cat_df
```

```python
# Creat database connection
connection_string = "postgres:postgres@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
# Add category data to sql
cat_df.to_sql(name='categories', con=engine, if_exists='append', index=True)
```

```python
# Add video data to sql
us_videos.to_sql(name='videos', con=engine, if_exists='append', index=True)
```
